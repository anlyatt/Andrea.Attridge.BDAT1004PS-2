Question 1: What value is displayed when the last expression (a) is evaluated? Explain your 
answer by indicating what happens in every executed statement.

```python
a = 0

def b():
    global a
    a = c(a)

def c(a):
    return a + 2

print(b())
print(b())
print(b())
print(a)
```

Question 2: Function fileLength(), given to you, takes the name of a file as input and returns the length of the file.  Modify function fileLength() so that a friendly message is printed instead:

```python
def file_length(file_name):
    
    try:
        #try block
        file = open(file_name)
        contents = file.read()
        file.close()
        print(len(contents))
    except FileNotFoundError:
        #except block - executed only if a FileNotFoundError
        #exception is reasied in the try block
        print ('File name {} not found'.format(file_name))
    
file_length('midterm.py.txt')

file_length('idterm.py.txt')
```

```python
def collatz(n):
    
    #prints the initial value of n
    print(n)
    
    #if n equals 1, stops
    if n == 1:
        return
    
    #if n is even, prints the result of n divided by 2 and creates that as the new value of n
    elif n % 2 == 0:
        collatz(n//2)
    
    #if n is odd, prints the result of n times 3 plus 1 and creates that as the new value of n
    else:
        collatz (3*n + 1)

print('Case one (n=1):')
collatz(1)

print('\nCase two (n=10):')
collatz(10)
        
```

Question 5: Write a recursive method binary() that takes a non-negative integer n and prints the binary representation of integer n.

```python
def binary(n):
    
    #if n is greater than 1, divide it by 2 - us thei is the new value of n for the function
    if n > 1:
        binary(n // 2)
    
    #print the remainder of n divided by 2
    print(n % 2, end='')

print('Case one (n=0):')
binary(0)
print('\nCase two (n=1):')
binary(1)
print('\nCase three (n=3):')
binary(3)
print('\nCase four (n=9):')
binary (9)
```

Question 6: Implement a class named HeadingParser that can be used to parse an HTML document, and retrieve and print all the headings in the document. You should implement your class as a subclass of HTMLParser, defined in Standard Library 
module html.parser. When fed a string containing HTML code, your class should print the headings, one per line and in the order in which they appear in the document. Each heading should be indented as follows: an h1 heading should have indentation 0, and h2 heading should have indentation 1, etc. Test your implementation using w3c.html.

```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    
    def __init__(self):
        'initializes the parser and the initial indentation'
        super().__init__()
        self.indent = 0
        self.current_line = 1
        self.current_tag = None
        
    def handle_starttag(self, tag, attrs):
        self.current_tag = tag
        self.indent += 4 if tag == 'h1' else 8 if tag == 'h2' else 0

    def handle_endtag(self, tag):
        self.indent -= 4 if tag == 'h1' else 8 if tag == 'h2' else 0
        self.current_tag = None
        
    def handle_data(self, data):
        if self.current_tag in {'h1', 'h2'}:
            print('{}{}'.format(self.indent * ' ', data.strip()))
        
hp = HeadingParser()

with open('w3c.txt', 'r') as file:
    for line in file:
        hp.current_line += 1
        hp.feed(line)
```

Question 7: Implement recursive function webdir() that takes as input: a URL (as a string) and non-negative integers depth and indent. Your function should visit every web page reachable from the starting URL web page in depth clicks or less, and print each web page's URL. As shown below, indentation, specified by indent, should be used to indicate the depth of a URL.

```python
#import requirements
from urllib.request import urlopen
from urllib.parse import urljoin
from html.parser import HTMLParser

class Collector(HTMLParser):
    'collects hyperlink URLs into a list'

    def __init__(self, url):
        'initializes parser, the url, and a list'
        HTMLParser.__init__(self)
        self.url = url
        self.links = []


    def handle_starttag(self, tag, attrs):
        'collects hyperlink URLs in their absolute format'
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    # construct absolute URL
                    absolute = urljoin(self.url, attr[1])
                    if absolute[:4] == 'http': # collect HTTP URLs
                        self.links.append(absolute)
                        
    def getLinks(self):
        'returns hyperlinks URLs in their absolute format'
        return self.links

def analyze(url, depth, indent):
    
    print('{}{:45}'.format(' ' * indent, url))  # print the current URL with indentation

    # obtain links in the web page
    content = urlopen(url).read().decode()
    collector = Collector(url)
    collector.feed(content)
    urls = collector.getLinks()          # get list of links

    for link in urls:
        print('{}{:45}'.format(' ' * (indent + 2), link))

    return urls

visited = set() # initialize visited to an empty set

def webdir(url, depth, indent):
    
    # add url to set of visited pages
    global visited     # warns the programmer 
    visited.add(url)

    # analyze() returns a list of hyperlink URLs in web page url 
    links = analyze(url, depth, indent)

    # recursively continue crawl from every link in links
    if depth > 0:
        for link in links:
            # follow link only if not visited
            if link not in visited:
                try:
                    webdir(link, depth - 1, indent + 2)
                except:
                    pass

webdir('https://www.boredpanda.com/', 2, 2)
```
